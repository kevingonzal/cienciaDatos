```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt  
import seaborn as sns 

from google.colab import drive
drive.mount('/content/gdrive')
```

```python
df = pd.read_excel('/content/drive/MyDrive/bloque ciencia datos/alphabet_stock_data.xlsx')
```

```python
df['Date'] = pd.to_datetime(df['Date'])

# seleccionar los datos entre las dos fechas concretas
fecha_inicio = pd.to_datetime('20/04/20')
fecha_fin = pd.to_datetime('30/04/20')


df_filtrado = df[(df['Date'] >= fecha_inicio) & (df['Date'] <= fecha_fin)]

# crear el gráfico de líneas utilizando Matplotlib
plt.plot(df_filtrado['Date'], df_filtrado['Close'])
plt.title('Precios Históricos entre 2020-04-20 y 2020-04-30')
plt.xlabel('Fecha')
plt.ylabel('Precio')
plt.show()
```
