```python
import numpy as np
import math
from simpleai.search import SearchProblem, breadth_first
from simpleai.search.viewers import BaseViewer
from simpleai.search import astar, depth_first, uniform_cost, greedy


mars_map = np.load('mars_map1.npy')
nr, nc = mars_map.shape
scale = 10.0174
Heightparameter = 0.25

def scaleX(x):
    Q = round(x/scale)
    return Q

def scaleY(y):
    W = nr-round(y/scale)
    return W

def Parametertest(W, Q, nr, nc):
    return abs(mars_map[W][Q] - mars_map[nr][nc]) < Heightparameter

def distance(coordinateone, coordinatetwo):
    y1, x1 = coordinateone
    y2, x2 = coordinatetwo
    return math.sqrt((y2 - y1)**2 + (x2 - x1)**2)

class Marsroute(SearchProblem):
    def __init__(self):
        Realstate = (scaleY(6400), scaleX(2850))
        initial_state = Realstate
        SearchProblem.__init__(self, initial_state)
        Realstatef = (scaleY(6800), scaleX(3150))
        self.goal_state = Realstatef

    def actions(self, state):
        actions = []
        W = state[0]
        Q = state[1]
        newW = W-1
        newQ = Q-1
        if W != 0 and Q != 0 and (Parametertest(W, Q, newW, newQ) == True) and (mars_map[newW][newQ] > -1):
            actions.append((W, Q, newW, newQ))
        newW = W-1
        newQ = Q
        if W != 0 and (Parametertest(W, Q, newW, newQ)) == True and (mars_map[newW][newQ] > -1):
            actions.append((W, Q, newW, newQ))
        newW = W - 1
        newQ = Q + 1
        if W != 0 and Q != nc-1 and (Parametertest(W, Q, newW, newQ) == True) and (mars_map[newW][newQ] > -1):
            actions.append((W, Q, newW, newQ))
        newW = W
        newQ = Q-1
        if Q != 0 and (Parametertest(W, Q, newW, newQ) == True) and (mars_map[newW][newQ] > -1):
            actions.append((W, Q, newW, newQ))
        newW = W
        newQ = Q+1
        if Q != nc-1 and (Parametertest(W, Q, newW, newQ) == True) and (mars_map[newW][newQ] > -1):
            actions.append((W, Q, newW, newQ))
        newW = W+1
        newQ = Q-1
        if W != nr-1 and Q != 0 and (Parametertest(W, Q, newW, newQ) == True) and (mars_map[newW][newQ] > -1):
            actions.append((W, Q, newW, newQ))
        newW = W-1
        newQ = Q
        if W != nc-1 and (Parametertest(W, Q, newW, newQ) == True) and (mars_map[newW][newQ] > -1):
            actions.append((W, Q, newW, newQ))
        newW = W+1
        newQ = Q+1
        if W != nr-1 and Q != nc-1 and (Parametertest(W, Q, newW, newQ) == True) and (mars_map[newW][newQ] > -1):
            actions.append((W, Q, newW, newQ))

        return actions

    def result(self, state, action):
        new_state = (action[2], action[3])
        return new_state

    def is_goal(self, state):
        found_goal = state == self.goal_state
        return found_goal

    def cost(self, state, action, state_two):
        return distance(state, state_two) * scale

    def heuristic(self, state):
        return distance(state, self.goal_state)

def display(result):
    if result is not None:
        for i, (action, state) in enumerate(result.path()):
            if action is None:
                print('Configuración inicial')
            elif i == len(result.path()) - 1:
                print(i, '- Después de (', action[0], ',', action[1], ')', 'se movió a (', action[2], ',', action[3], ')')
                print('¡La distancia recorrida por el robot: ', result.cost, 'en metros!')
            else:
                print(i, '- Después de (', action[0], ',', action[1], ')', 'se movió a (', action[2], ',', action[3], ')')
            print('  ', state)
    else:
        print('Mala configuración del problema')
```
