# Model for TEXT

```python
df = pd.DataFrame()
for i in range(n):
    # TEXT
    data=pd.read_csv(text_data[i])
    target=pd.read_csv(target_vectors_text[i])
    data = data.head(len(target))
    data['Score class'] = target['Score class']
    
    df = df.append(data, ignore_index=True)
df    
```

```python
X = df.iloc[:, :-1]  
y = df.iloc[:, -1]  

scaler = StandardScaler()
X_normalized = scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_normalized, y, test_size=0.2, random_state=42)

model_names = [
    'Logistic Regression',
    'K Nearest Neighbors',
    'Decision Tree',
    'Random Forest',
    'Support Vector Machine',
    'Gradient Boosting',
    'Neural Network (MLP)'
]

classifiers = [
    LogisticRegression(max_iter=1000),
    KNeighborsClassifier(),
    DecisionTreeClassifier(),
    RandomForestClassifier(),
    SVC(),
    GradientBoostingClassifier(),
    MLPClassifier(max_iter=1000)
]

results_df = pd.DataFrame(columns=['Model', 'Accuracy'])

for name, clf in zip(model_names, classifiers):
    clf.fit(X_train, y_train)
    accuracy = clf.score(X_test, y_test)
    results_df = results_df.append({'Model': name, 'Accuracy': accuracy}, ignore_index=True)

results_df.sort_values(by='Accuracy', ascending=False, inplace=True)
results_df
```

# Parameter Optimization for MLP

```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.neural_network import MLPClassifier

# Paso 1: Dividir el DataFrame en características (X) y etiquetas (y)
X = df.iloc[:, :-1]  
y = df.iloc[:, -1]  
# Paso 2: Dividir los datos en conjuntos de entrenamiento y prueba
scaler = StandardScaler()
X_normalized = scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_normalized, y, test_size=0.2, random_state=42)

# Paso 3: Definir una cuadrícula de parámetros a ser explorados
param_grid = {
    'hidden_layer_sizes': [(50,), (100,), (50, 50)],      # Número de neuronas en cada capa oculta
    'activation': ['logistic', 'relu'],                  # Función de activación ('logistic' para sigmoide, 'relu' para ReLU)
    'solver': ['adam', 'sgd'],                           # Algoritmo de optimización ('adam' para Adam, 'sgd' para descenso de gradiente estocástico)
    'alpha': [0.0001, 0.001, 0.01],                      # Término de regularización L2
    'max_iter': [100, 500, 1000]                         # Número máximo de iteraciones
}

# Paso 4: Utilizar validación cruzada con búsqueda en cuadrícula para encontrar los mejores parámetros
mlp = MLPClassifier()
grid_search = GridSearchCV(mlp, param_grid, cv=5, n_jobs=-1)
grid_search.fit(X_train, y_train)

# Mejores parámetros encontrados
best_params = grid_search.best_params_
print("Mejores parámetros:", best_params)

# Paso 5: Entrenar el modelo con los mejores parámetros encontrados
best_mlp = MLPClassifier(**best_params)
best_mlp.fit(X_train, y_train)

# Evaluar el modelo optimizado en el conjunto de prueba
accuracy = best_mlp.score(X_test, y_test)
print("Accuracy con parámetros optimizados:", accuracy)
```
